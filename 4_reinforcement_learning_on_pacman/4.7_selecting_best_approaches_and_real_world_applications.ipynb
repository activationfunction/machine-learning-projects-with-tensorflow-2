## Easy Agents

```python
from easyagents.agents import DqnAgent # pip install easyagents, [pip uninstall -y enum34]
from easyagents.agents import PpoAgent
from easyagents.agents import EasyAgent
from easyagents.callbacks import plot
```

```python
ppoAgent = DqnAgent('MsPacman-ram-v0', fc_layers=(200, 55))
ppoAgent.train([plot.State(), plot.Loss(), plot.Actions(), plot.Rewards()], 
               num_iterations=4000, num_iterations_between_eval=10)
```

```python
ppoAgent.play([plot.State(), plot.Actions(), plot.StepRewards()], 
              num_episodes=5, 
              max_steps_per_episode=1000)
```
